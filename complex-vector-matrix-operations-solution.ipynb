## Complex Column Vector Operations

### Exercise 1: Sum of Two Complex Vectors

Given two complex vectors:
- $v_1 = \begin{bmatrix} 1 + 2i \\ 3 - i \end{bmatrix}$
- $v_2 = \begin{bmatrix} 2 - i \\ 4 + 2i \end{bmatrix}$

Find their sum.

```python
# Exercise 1: Sum of Two Complex Vectors
print("=== Exercise 1: Sum of Two Complex Vectors ===")

# Define the complex vectors
v1 = np.array([[1 + 2j],
               [3 - 1j]])

v2 = np.array([[2 - 1j],
               [4 + 2j]])

# Calculate the sum
v_sum = v1 + v2

print("Vector v1:")
print(v1)
print("\nVector v2:")
print(v2)
print("\nSum v1 + v2:")
print(v_sum)

# Verification: manually calculate
print("\nManual verification:")
print(f"Element 1: (1+2j) + (2-1j) = {(1+2j) + (2-1j)}")
print(f"Element 2: (3-1j) + (4+2j) = {(3-1j) + (4+2j)}")
```

### Exercise 2: Negation of a Complex Vector

Given a complex vector:
- $v = \begin{bmatrix} 2 + 3i \\ 1 - 2i \end{bmatrix}$

Find its negation.

```python
# Exercise 2: Negation of a Complex Vector
print("=== Exercise 2: Negation of a Complex Vector ===")

# Define the complex vector
v = np.array([[2 + 3j],
              [1 - 2j]])

# Calculate the negation
v_neg = -v

print("Original vector v:")
print(v)
print("\nNegation -v:")
print(v_neg)

print("\nManual verification:")
print(f"Element 1: -(2+3j) = {-(2+3j)}")
print(f"Element 2: -(1-2j) = {-(1-2j)}")
```

### Exercise 3: Multiplication of a Complex Vector by a Scalar

Multiply the complex vector:
- $v = \begin{bmatrix} 1 - i \\ 2 + 2i \end{bmatrix}$

by the complex scalar $3 + i$.

```python
# Exercise 3: Multiplication of a Complex Vector by a Scalar
print("=== Exercise 3: Scalar Multiplication ===")

# Define the complex vector and scalar
v = np.array([[1 - 1j],
              [2 + 2j]])

scalar = 3 + 1j

# Calculate scalar multiplication
v_scaled = v * scalar

print(f"Vector v:")
print(v)
print(f"\nScalar: {scalar}")
print("\nResult of scalar multiplication:")
print(v_scaled)

print("\nManual verification:")
print(f"Element 1: (1-1j) * (3+1j) = {(1-1j) * (3+1j)}")
print(f"Element 2: (2+2j) * (3+1j) = {(2+2j) * (3+1j)}")

# Show the calculation step by step for first element
print("\nStep-by-step for first element:")
print("(1-j)(3+j) = 1*3 + 1*j + (-j)*3 + (-j)*j")
print("           = 3 + j - 3j - j²")
print("           = 3 + j - 3j - (-1)")
print("           = 3 + j - 3j + 1")
print("           = 4 - 2j")
```

### Exercise 4: Transpose of a Complex Vector

Find the transpose of the complex vector:
- $v = \begin{bmatrix} 2 - i \\ 3 + 4i \end{bmatrix}$

```python
# Exercise 4: Transpose of a Complex Vector
print("=== Exercise 4: Transpose of a Complex Vector ===")

# Define the complex vector
v = np.array([[2 - 1j],
              [3 + 4j]])

# Calculate the transpose
v_transpose = v.T

print("Original vector v (column vector):")
print(v)
print(f"Shape: {v.shape}")

print("\nTranspose v^T (row vector):")
print(v_transpose)
print(f"Shape: {v_transpose.shape}")

print("\nNote: Transpose converts column vector to row vector")
```

### Exercise 5: Conjugate of a Complex Vector

Compute the conjugate of the complex vector:
- $v = \begin{bmatrix} 1 + i \\ 2 - 3i \end{bmatrix}$

```python
# Exercise 5: Conjugate of a Complex Vector
print("=== Exercise 5: Conjugate of a Complex Vector ===")

# Define the complex vector
v = np.array([[1 + 1j],
              [2 - 3j]])

# Calculate the conjugate
v_conjugate = np.conjugate(v)

print("Original vector v:")
print(v)

print("\nConjugate v*:")
print(v_conjugate)

print("\nManual verification:")
print(f"Element 1: conjugate(1+1j) = {np.conjugate(1+1j)}")
print(f"Element 2: conjugate(2-3j) = {np.conjugate(2-3j)}")

print("\nNote: Conjugate changes the sign of the imaginary part")
```

### Exercise 6: Adjunct (Conjugate Transpose) of a Complex Vector

Find the adjunct (conjugate transpose) of the complex vector:
- $v = \begin{bmatrix} 1 - 2i \\ 3 + i \end{bmatrix}$

```python
# Exercise 6: Adjunct (Conjugate Transpose) of a Complex Vector
print("=== Exercise 6: Adjunct (Conjugate Transpose) ===")

# Define the complex vector
v = np.array([[1 - 2j],
              [3 + 1j]])

# Calculate the adjunct (conjugate transpose)
v_adjunct = v.T.conj()
# Alternative way: v_adjunct = np.conj(v.T)

print("Original vector v:")
print(v)
print(f"Shape: {v.shape}")

print("\nTranspose v^T:")
print(v.T)

print("\nConjugate of v:")
print(np.conj(v))

print("\nAdjunct v† (conjugate transpose):")
print(v_adjunct)
print(f"Shape: {v_adjunct.shape}")

print("\nVerification: v† = (v^T)*")
print("The adjunct is the transpose of the conjugate, or equivalently, the conjugate of the transpose")
```

## Complex Square Matrix Operations

### Exercise 7: Sum of Two Complex Matrices

Given two complex matrices:
- $m_1 = \begin{bmatrix} 1 + i & 2 - i \\ 3 + 2i & 4 \end{bmatrix}$
- $m_2 = \begin{bmatrix} 2 - 3i & 1 \\ i & 2 + 2i \end{bmatrix}$

Calculate their sum.

```python
# Exercise 7: Sum of Two Complex Matrices
print("=== Exercise 7: Sum of Two Complex Matrices ===")

# Define the complex matrices
m1 = np.array([[1 + 1j, 2 - 1j],
               [3 + 2j, 4 + 0j]])

m2 = np.array([[2 - 3j, 1 + 0j],
               [0 + 1j, 2 + 2j]])

# Calculate the sum
m_sum = m1 + m2

print("Matrix m1:")
print(m1)

print("\nMatrix m2:")
print(m2)

print("\nSum m1 + m2:")
print(m_sum)

print("\nElement-wise verification:")
print(f"(0,0): (1+1j) + (2-3j) = {(1+1j) + (2-3j)}")
print(f"(0,1): (2-1j) + (1+0j) = {(2-1j) + (1+0j)}")
print(f"(1,0): (3+2j) + (0+1j) = {(3+2j) + (0+1j)}")
print(f"(1,1): (4+0j) + (2+2j) = {(4+0j) + (2+2j)}")
```

### Exercise 8: Negation of a Complex Matrix

Find the negation of the complex matrix:
- $m = \begin{bmatrix} 2 + i & 3 \\ 1 - i & 2 + 2i \end{bmatrix}$

```python
# Exercise 8: Negation of a Complex Matrix
print("=== Exercise 8: Negation of a Complex Matrix ===")

# Define the complex matrix
m = np.array([[2 + 1j, 3 + 0j],
              [1 - 1j, 2 + 2j]])

# Calculate the negation
m_neg = -m

print("Original matrix m:")
print(m)

print("\nNegation -m:")
print(m_neg)

print("\nElement-wise verification:")
print(f"(0,0): -(2+1j) = {-(2+1j)}")
print(f"(0,1): -(3+0j) = {-(3+0j)}")
print(f"(1,0): -(1-1j) = {-(1-1j)}")
print(f"(1,1): -(2+2j) = {-(2+2j)}")
```

### Exercise 9: Multiplication of a Complex Matrix by a Scalar

Multiply the complex matrix:
- $m = \begin{bmatrix} 1 - i & 2 \\ 3 + i & 4 - 2i \end{bmatrix}$

by the complex scalar $2 + 3i$.

```python
# Exercise 9: Multiplication of a Complex Matrix by a Scalar
print("=== Exercise 9: Scalar Multiplication of Matrix ===")

# Define the complex matrix and scalar
m = np.array([[1 - 1j, 2 + 0j],
              [3 + 1j, 4 - 2j]])

scalar = 2 + 3j

# Calculate scalar multiplication
m_scaled = m * scalar

print("Matrix m:")
print(m)

print(f"\nScalar: {scalar}")

print("\nResult of scalar multiplication:")
print(m_scaled)

print("\nElement-wise verification:")
print(f"(0,0): (1-1j) * (2+3j) = {(1-1j) * (2+3j)}")
print(f"(0,1): (2+0j) * (2+3j) = {(2+0j) * (2+3j)}")
print(f"(1,0): (3+1j) * (2+3j) = {(3+1j) * (2+3j)}")
print(f"(1,1): (4-2j) * (2+3j) = {(4-2j) * (2+3j)}")

# Show detailed calculation for one element
print("\nDetailed calculation for (0,0):")
print("(1-j)(2+3j) = 1*2 + 1*3j + (-j)*2 + (-j)*3j")
print("            = 2 + 3j - 2j - 3j²")
print("            = 2 + 3j - 2j - 3(-1)")
print("            = 2 + 3j - 2j + 3")
print("            = 5 + j")
```

### Exercise 10: Conjugate of a Complex Matrix

Compute the conjugate of the complex matrix:
- $m = \begin{bmatrix} 1 + i & 2 \\ 3 - i & 4 + 2i \end{bmatrix}$

```python
# Exercise 10: Conjugate of a Complex Matrix
print("=== Exercise 10: Conjugate of a Complex Matrix ===")

# Define the complex matrix
m = np.array([[1 + 1j, 2 + 0j],
              [3 - 1j, 4 + 2j]])

# Calculate the conjugate
m_conjugate = np.conjugate(m)

print("Original matrix m:")
print(m)

print("\nConjugate m*:")
print(m_conjugate)

print("\nElement-wise verification:")
print(f"(0,0): conjugate(1+1j) = {np.conjugate(1+1j)}")
print(f"(0,1): conjugate(2+0j) = {np.conjugate(2+0j)}")
print(f"(1,0): conjugate(3-1j) = {np.conjugate(3-1j)}")
print(f"(1,1): conjugate(4+2j) = {np.conjugate(4+2j)}")
```

### Exercise 11: Transpose of a Complex Matrix

Find the transpose of the complex matrix:
- $m = \begin{bmatrix} 1 - i & 2 + 2i \\ 3 & 4 - i \end{bmatrix}$

```python
# Exercise 11: Transpose of a Complex Matrix
print("=== Exercise 11: Transpose of a Complex Matrix ===")

# Define the complex matrix
m = np.array([[1 - 1j, 2 + 2j],
              [3 + 0j, 4 - 1j]])

# Calculate the transpose
m_transpose = m.T

print("Original matrix m:")
print(m)

print("\nTranspose m^T:")
print(m_transpose)

print("\nVerification - element mapping:")
print("m[0,0] = 1-1j → m^T[0,0] = 1-1j")
print("m[0,1] = 2+2j → m^T[1,0] = 2+2j")
print("m[1,0] = 3+0j → m^T[0,1] = 3+0j")
print("m[1,1] = 4-1j → m^T[1,1] = 4-1j")

print("\nNote: Transpose swaps rows and columns: m^T[i,j] = m[j,i]")
```

### Exercise 12: Adjunct (Conjugate Transpose) of a Complex Matrix

Calculate the adjunct (conjugate transpose) of the complex matrix:
- $m = \begin{bmatrix} 1 + 2i & 3 - i \\ 4 & 5 + i \end{bmatrix}$

```python
# Exercise 12: Adjunct (Conjugate Transpose) of a Complex Matrix
print("=== Exercise 12: Adjunct (Conjugate Transpose) ===")

# Define the complex matrix
m = np.array([[1 + 2j, 3 - 1j],
              [4 + 0j, 5 + 1j]])

# Calculate the adjunct (conjugate transpose)
m_adjunct = m.T.conj()

print("Original matrix m:")
print(m)

print("\nTranspose m^T:")
print(m.T)

print("\nConjugate m*:")
print(np.conj(m))

print("\nAdjunct m† (conjugate transpose):")
print(m_adjunct)

print("\nStep-by-step verification:")
print("1. Original: m[0,0] = 1+2j → Transpose: m^T[0,0] = 1+2j → Adjunct: m†[0,0] = 1-2j")
print("2. Original: m[0,1] = 3-1j → Transpose: m^T[1,0] = 3-1j → Adjunct: m†[1,0] = 3+1j")
print("3. Original: m[1,0] = 4+0j → Transpose: m^T[0,1] = 4+0j → Adjunct: m†[0,1] = 4-0j")
print("4. Original: m[1,1] = 5+1j → Transpose: m^T[1,1] = 5+1j → Adjunct: m†[1,1] = 5-1j")

print("\nNote: The adjunct (Hermitian transpose) is crucial in quantum mechanics and linear algebra")
```

### Exercise 13: Checking the Dimensions of a Matrix

Determine the dimensions of the complex matrix:
- $m = \begin{bmatrix} 1 - i & 2 \\ 3 + 2i & 4 - i \end{bmatrix}$

```python
# Exercise 13: Checking the Dimensions of a Matrix
print("=== Exercise 13: Matrix Dimensions ===")

# Define the complex matrix
m = np.array([[1 - 1j, 2 + 0j],
              [3 + 2j, 4 - 1j]])

print("Matrix m:")
print(m)

# Get various dimension information
shape = m.shape
rows, cols = m.shape
size = m.size
ndim = m.ndim

print(f"\nDimension information:")
print(f"Shape: {shape}")
print(f"Number of rows: {rows}")
print(f"Number of columns: {cols}")
print(f"Total number of elements: {size}")
print(f"Number of dimensions: {ndim}")

print(f"\nMatrix type: {rows}×{cols} {'square' if rows == cols else 'rectangular'} matrix")

# Additional information
print(f"\nAdditional information:")
print(f"Data type: {m.dtype}")
print(f"Memory layout: {'C-contiguous' if m.flags['C_CONTIGUOUS'] else 'Not C-contiguous'}")
```

## Multiplication of Matrices and Vectors

### Exercise 14: Multiplying a Complex Square Matrix by a Complex Column Vector

Multiply the complex square matrix:
- $m = \begin{bmatrix} 1 + i & 2 - i \\ 3 & 4 + i \end{bmatrix}$

by the complex column vector:
- $v = \begin{bmatrix} 2 - i \\ 1 + 3i \end{bmatrix}$

```python
# Exercise 14: Matrix-Vector Multiplication
print("=== Exercise 14: Matrix-Vector Multiplication ===")

# Define the complex matrix and vector
m = np.array([[1 + 1j, 2 - 1j],
              [3 + 0j, 4 + 1j]])

v = np.array([[2 - 1j],
              [1 + 3j]])

# Perform matrix-vector multiplication
result = np.dot(m, v)
# Alternative: result = m @ v

print("Matrix m:")
print(m)
print(f"Shape: {m.shape}")

print("\nVector v:")
print(v)
print(f"Shape: {v.shape}")

print("\nResult of m × v:")
print(result)
print(f"Shape: {result.shape}")

# Manual verification
print("\nManual calculation:")
print("Result[0] = m[0,0]*v[0] + m[0,1]*v[1]")
print(f"         = (1+1j)*(2-1j) + (2-1j)*(1+3j)")
element_0 = (1+1j)*(2-1j) + (2-1j)*(1+3j)
print(f"         = {(1+1j)*(2-1j)} + {(2-1j)*(1+3j)}")
print(f"         = {element_0}")

print("\nResult[1] = m[1,0]*v[0] + m[1,1]*v[1]")
print(f"         = (3+0j)*(2-1j) + (4+1j)*(1+3j)")
element_1 = (3+0j)*(2-1j) + (4+1j)*(1+3j)
print(f"         = {(3+0j)*(2-1j)} + {(4+1j)*(1+3j)}")
print(f"         = {element_1}")

# Detailed calculation for first element
print("\nDetailed calculation for first element:")
print("(1+j)(2-j) = 1*2 + 1*(-j) + j*2 + j*(-j)")
print("           = 2 - j + 2j - j²")
print("           = 2 - j + 2j - (-1)")
print("           = 2 + j + 1 = 3 + j")

print("(2-j)(1+3j) = 2*1 + 2*3j + (-j)*1 + (-j)*3j")
print("            = 2 + 6j - j - 3j²")
print("            = 2 + 6j - j - 3(-1)")
print("            = 2 + 6j - j + 3 = 5 + 5j")

print("Total: (3+j) + (5+5j) = 8 + 6j")
```

## Bonus: Advanced Operations and Visualizations

```python
# Bonus: Visualization of Complex Matrix Operations
print("=== Bonus: Complex Matrix Visualizations ===")

# Create a sample complex matrix
complex_matrix = np.array([[1+2j, 3-1j, 2+2j],
                          [2-1j, 1+3j, 1-1j],
                          [3+1j, 2-2j, 4+0j]])

# Visualize real and imaginary parts
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Real part
im1 = axes[0, 0].imshow(complex_matrix.real, cmap='RdBu', interpolation='nearest')
axes[0, 0].set_title('Real Part')
axes[0, 0].set_xlabel('Column')
axes[0, 0].set_ylabel('Row')
plt.colorbar(im1, ax=axes[0, 0])

# Imaginary part
im2 = axes[0, 1].imshow(complex_matrix.imag, cmap='RdBu', interpolation='nearest')
axes[0, 1].set_title('Imaginary Part')
axes[0, 1].set_xlabel('Column')
axes[0, 1].set_ylabel('Row')
plt.colorbar(im2, ax=axes[0, 1])

# Magnitude
im3 = axes[0, 2].imshow(np.abs(complex_matrix), cmap='viridis', interpolation='nearest')
axes[0, 2].set_title('Magnitude |z|')
axes[0, 2].set_xlabel('Column')
axes[0, 2].set_ylabel('Row')
plt.colorbar(im3, ax=axes[0, 2])

# Phase
im4 = axes[1, 0].imshow(np.angle(complex_matrix), cmap='hsv', interpolation='nearest')
axes[1, 0].set_title('Phase (angle)')
axes[1, 0].set_xlabel('Column')
axes[1, 0].set_ylabel('Row')
plt.colorbar(im4, ax=axes[1, 0])

# Transpose
im5 = axes[1, 1].imshow(complex_matrix.T.real, cmap='RdBu', interpolation='nearest')
axes[1, 1].set_title('Transpose (Real Part)')
axes[1, 1].set_xlabel('Column')
axes[1, 1].set_ylabel('Row')
plt.colorbar(im5, ax=axes[1, 1])

# Conjugate
im6 = axes[1, 2].imshow(np.conj(complex_matrix).imag, cmap='RdBu', interpolation='nearest')
axes[1, 2].set_title('Conjugate (Imaginary Part)')
axes[1, 2].set_xlabel('Column')
axes[1, 2].set_ylabel('Row')
plt.colorbar(im6, ax=axes[1, 2])

plt.tight_layout()
plt.show()

# Summary statistics
print("\nMatrix Statistics:")
print(f"Original matrix shape: {complex_matrix.shape}")
print(f"Maximum magnitude: {np.max(np.abs(complex_matrix)):.2f}")
print(f"Minimum magnitude: {np.min(np.abs(complex_matrix)):.2f}")
print(f"Mean real part: {np.mean(complex_matrix.real):.2f}")
print(f"Mean imaginary part: {np.mean(complex_matrix.imag):.2f}")
```

```python
# Bonus: Properties and Verification
print("=== Bonus: Matrix Properties Verification ===")

# Create test matrices
A = np.array([[1+2j, 3-1j], [2+1j, 4-2j]])
B = np.array([[2-1j, 1+1j], [3+2j, 1-3j]])

print("Testing matrix properties:")

# 1. (A + B)† = A† + B†
lhs = (A + B).T.conj()
rhs = A.T.conj() + B.T.conj()
prop1 = np.allclose(lhs, rhs)
print(f"1. (A + B)† = A† + B†: {prop1}")

# 2. (A†)† = A
prop2 = np.allclose(A.T.conj().T.conj(), A)
print(f"2. (A†)† = A: {prop2}")

# 3. (αA)† = α*A† for scalar α
alpha = 2 + 3j
lhs = (alpha * A).T.conj()
rhs = np.conj(alpha) * A.T.conj()
prop3 = np.allclose(lhs, rhs)
print(f"3. (αA)† = α*A†: {prop3}")

# 4. Check if a matrix is Hermitian (A = A†)
H = np.array([[1+0j, 2-3j], [2+3j, 4+0j]])  # Hermitian matrix
is_hermitian = np.allclose(H, H.T.conj())
print(f"4. H is Hermitian (H = H†): {is_hermitian}")

print("\nAll fundamental properties of complex matrices verified!")
```
