## Exercise 1: Basic Complex Number Operations

**Problem 1:** Manually calculate and then verify using Python the result of `(3 + 4j) + (1 - 2j)`, `(5 + 6j) * (7 - 8j)`, and `(2 - 3j) / (1 + 4j)`.

**Problem 2:** Find the conjugate and modulus of `(3 - 4j)`.

```python
# Exercise 1 Solutions

# Define complex numbers
a = 3 + 4j
b = 1 - 2j
c = 5 + 6j
d = 7 - 8j
e = 2 - 3j
f = 1 + 4j

print("=== Problem 1: Basic Operations ===")

# Addition: (3 + 4j) + (1 - 2j)
add_result = a + b
print(f"Addition: ({a}) + ({b}) = {add_result}")

# Multiplication: (5 + 6j) * (7 - 8j)
mult_result = c * d
print(f"Multiplication: ({c}) * ({d}) = {mult_result}")

# Manual calculation for multiplication:
# (5 + 6j)(7 - 8j) = 5*7 + 5*(-8j) + 6j*7 + 6j*(-8j)
# = 35 - 40j + 42j - 48j²
# = 35 + 2j - 48(-1) = 35 + 2j + 48 = 83 + 2j
print(f"Manual calculation: (5 + 6j)(7 - 8j) = 83 + 2j")

# Division: (2 - 3j) / (1 + 4j)
div_result = e / f
print(f"Division: ({e}) / ({f}) = {div_result}")

print("\n=== Problem 2: Conjugate and Modulus ===")

# Conjugate and modulus of (3 - 4j)
z = 3 - 4j
conjugate_z = z.conjugate()
modulus_z = abs(z)

print(f"Complex number: {z}")
print(f"Conjugate: {conjugate_z}")
print(f"Modulus: {modulus_z}")
print(f"Manual modulus calculation: sqrt(3² + (-4)²) = sqrt(9 + 16) = sqrt(25) = 5")
```

## Exercise 2: Visualizing Complex Operations

**Problem 1:** Write a Python function to plot a complex number on the complex plane. Use it to plot `(3 + 4j)` and its conjugate.

**Problem 2:** Extend the function to show addition and multiplication of two complex numbers graphically.

```python
# Exercise 2 Solutions

import matplotlib.pyplot as plt
import numpy as np

def plot_complex(z, label, color='blue', marker='o'):
    """Plot a complex number on the complex plane"""
    plt.plot(z.real, z.imag, marker, color=color, markersize=8, label=label)
    # Draw arrow from origin to the point
    plt.arrow(0, 0, z.real, z.imag, head_width=0.1, head_length=0.1, 
              fc=color, ec=color, alpha=0.5, length_includes_head=True)
    # Add text annotation
    plt.annotate(f'{z}', (z.real, z.imag), xytext=(5, 5), 
                textcoords='offset points', fontsize=10)

# Problem 1: Plot (3 + 4j) and its conjugate
plt.figure(figsize=(8, 6))
z1 = 3 + 4j
z1_conj = z1.conjugate()

plot_complex(z1, 'z = 3 + 4j', 'blue', 'o')
plot_complex(z1_conj, 'z* = 3 - 4j', 'red', 's')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, alpha=0.3)
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.title('Complex Number and its Conjugate')
plt.legend()
plt.axis('equal')
plt.show()

print("Problem 1: Plotted (3 + 4j) and its conjugate (3 - 4j)")
print("Notice how the conjugate is the reflection across the real axis")
```

```python
# Problem 2: Addition and multiplication visualization

# Define two complex numbers
z1 = 2 + 1j
z2 = 1 + 2j
z_add = z1 + z2
z_mult = z1 * z2

plt.figure(figsize=(12, 5))

# Addition visualization
plt.subplot(1, 2, 1)
plot_complex(z1, 'z₁ = 2 + j', 'blue', 'o')
plot_complex(z2, 'z₂ = 1 + 2j', 'green', 's')
plot_complex(z_add, 'z₁ + z₂ = 3 + 3j', 'red', '^')

# Show parallelogram rule for addition
plt.plot([z1.real, z_add.real], [z1.imag, z_add.imag], 'g--', alpha=0.5)
plt.plot([z2.real, z_add.real], [z2.imag, z_add.imag], 'b--', alpha=0.5)

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, alpha=0.3)
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.title('Complex Addition')
plt.legend()
plt.axis('equal')

# Multiplication visualization
plt.subplot(1, 2, 2)
plot_complex(z1, 'z₁ = 2 + j', 'blue', 'o')
plot_complex(z2, 'z₂ = 1 + 2j', 'green', 's')
plot_complex(z_mult, 'z₁ × z₂ = 0 + 5j', 'purple', 'D')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, alpha=0.3)
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.title('Complex Multiplication')
plt.legend()
plt.axis('equal')

plt.tight_layout()
plt.show()

print("Problem 2: Visualized addition and multiplication")
print(f"Addition: {z1} + {z2} = {z_add}")
print(f"Multiplication: {z1} × {z2} = {z_mult}")
print("Note: Multiplication rotates and scales the complex number")
```

## Exercise 3: Exploring the Mandelbrot Set

**Problem 1:** Modify the provided Mandelbrot set code to change its zoom level and center point. Observe how the fractal pattern changes.

**Problem 2:** Experiment with different values of `max_iter` and observe the effect on the fractal's detail and computation time.

```python
# Exercise 3 Solutions

import numpy as np
import matplotlib.pyplot as plt
import time

def mandelbrot(c, max_iter):
    """Compute the number of iterations for the Mandelbrot set"""
    z = 0
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + c
    return max_iter

def plot_mandelbrot(x_center, y_center, zoom, max_iter, pixels=400):
    """Plot Mandelbrot set with given parameters"""
    x_width = 3.0 / zoom
    y_width = 3.0 / zoom
    
    x = np.linspace(x_center - x_width, x_center + x_width, pixels)
    y = np.linspace(y_center - y_width, y_center + y_width, pixels)
    xx, yy = np.meshgrid(x, y)
    c = xx + yy*1j
    
    start_time = time.time()
    mandelbrot_set = np.array([mandelbrot(ci, max_iter) for ci in c.ravel()]).reshape(c.shape)
    end_time = time.time()
    
    plt.imshow(mandelbrot_set, extent=[x.min(), x.max(), y.min(), y.max()], 
               cmap='hot', interpolation='bilinear')
    plt.colorbar(label='Iterations to escape')
    plt.title(f'Mandelbrot Set (center={x_center}+{y_center}j, zoom={zoom}, iter={max_iter})')
    plt.xlabel('Real')
    plt.ylabel('Imaginary')
    
    return end_time - start_time

# Problem 1: Different zoom levels and center points
plt.figure(figsize=(15, 10))

# Original view
plt.subplot(2, 3, 1)
time1 = plot_mandelbrot(0, 0, 1, 50)
plt.title(f'Original View\nTime: {time1:.2f}s')

# Zoomed in on interesting region
plt.subplot(2, 3, 2)
time2 = plot_mandelbrot(-0.7, 0, 3, 50)
plt.title(f'Zoomed (3x)\nTime: {time2:.2f}s')

# Highly zoomed
plt.subplot(2, 3, 3)
time3 = plot_mandelbrot(-0.8, 0.156, 10, 50)
plt.title(f'Highly Zoomed (10x)\nTime: {time3:.2f}s')

# Problem 2: Different iteration counts
plt.subplot(2, 3, 4)
time4 = plot_mandelbrot(-0.7, 0, 3, 20)
plt.title(f'20 iterations\nTime: {time4:.2f}s')

plt.subplot(2, 3, 5)
time5 = plot_mandelbrot(-0.7, 0, 3, 100)
plt.title(f'100 iterations\nTime: {time5:.2f}s')

plt.subplot(2, 3, 6)
time6 = plot_mandelbrot(-0.7, 0, 3, 200)
plt.title(f'200 iterations\nTime: {time6:.2f}s')

plt.tight_layout()
plt.show()

print("Exercise 3 Analysis:")
print("1. Zoom Effect: Higher zoom reveals more fractal detail at the boundary")
print("2. Iteration Effect: More iterations reveal finer detail but take longer to compute")
print(f"3. Computation times: 20 iter: {time4:.2f}s, 100 iter: {time5:.2f}s, 200 iter: {time6:.2f}s")
```

## Exercise 4: Creating a Julia Set

**Problem 1:** Implement a Julia set generator. Use a constant like `-0.4 + 0.6j` for the iteration `z = z*z + constant`.

**Problem 2:** Explore how changing the constant changes the pattern of the Julia set.

```python
# Exercise 4 Solutions

def julia(z, constant, max_iter):
    """Compute the Julia set iteration"""
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + constant
    return max_iter

def plot_julia(constant, max_iter=100, pixels=400):
    """Plot Julia set for given constant"""
    x = np.linspace(-2, 2, pixels)
    y = np.linspace(-2, 2, pixels)
    xx, yy = np.meshgrid(x, y)
    z = xx + yy*1j
    
    julia_set = np.array([julia(zi, constant, max_iter) for zi in z.ravel()]).reshape(z.shape)
    
    plt.imshow(julia_set, extent=[x.min(), x.max(), y.min(), y.max()], 
               cmap='hot', interpolation='bilinear')
    plt.colorbar(label='Iterations to escape')
    plt.title(f'Julia Set (c = {constant})')
    plt.xlabel('Real')
    plt.ylabel('Imaginary')

# Problem 1 & 2: Different constants for Julia sets
constants = [
    -0.4 + 0.6j,     # Original constant
    -0.8 + 0.156j,   # Creates interesting spiral patterns
    -0.123 + 0.745j, # Creates branching patterns
    0.285 + 0.01j,   # Nearly real constant
    -0.75 + 0j,      # Real constant
    -0.1 + 0.8j      # High imaginary component
]

plt.figure(figsize=(18, 12))

for i, constant in enumerate(constants):
    plt.subplot(2, 3, i+1)
    plot_julia(constant, max_iter=100, pixels=300)

plt.tight_layout()
plt.show()

print("Exercise 4 Analysis:")
print("Julia sets show how different constants create vastly different fractal patterns:")
print("- Real constants tend to create more symmetric patterns")
print("- Complex constants with larger imaginary parts create more chaotic patterns")
print("- The Julia set is connected when the constant is in the Mandelbrot set")
print("- Each point in the Mandelbrot set corresponds to a different Julia set pattern")
```

## Exercise 5: Complex Number Properties

**Problem 1:** Prove that the absolute value of a product of two complex numbers is the product of their absolute values.

**Problem 2:** Show that the conjugate of a sum of two complex numbers is the sum of their conjugates.

```python
# Exercise 5 Solutions

print("=== Problem 1: |z₁ × z₂| = |z₁| × |z₂| ===")
print()

# Define test complex numbers
z1 = 3 + 4j
z2 = 1 - 2j

# Calculate product
product = z1 * z2

# Calculate absolute values
abs_product = abs(product)
abs_z1_times_abs_z2 = abs(z1) * abs(z2)

print(f"z₁ = {z1}")
print(f"z₂ = {z2}")
print(f"z₁ × z₂ = {product}")
print()
print(f"|z₁ × z₂| = {abs_product}")
print(f"|z₁| × |z₂| = {abs(z1)} × {abs(z2)} = {abs_z1_times_abs_z2}")
print(f"Equal? {np.isclose(abs_product, abs_z1_times_abs_z2)}")

print()
print("Mathematical proof:")
print("Let z₁ = a + bj and z₂ = c + dj")
print("z₁ × z₂ = (ac - bd) + (ad + bc)j")
print("|z₁ × z₂| = √[(ac - bd)² + (ad + bc)²]")
print("         = √[a²c² - 2abcd + b²d² + a²d² + 2abcd + b²c²]")
print("         = √[a²c² + b²d² + a²d² + b²c²]")
print("         = √[a²(c² + d²) + b²(c² + d²)]")
print("         = √[(a² + b²)(c² + d²)]")
print("         = √(a² + b²) × √(c² + d²)")
print("         = |z₁| × |z₂|")

print("\n" + "="*50)
print("=== Problem 2: (z₁ + z₂)* = z₁* + z₂* ===")
print()

# Calculate sum and its conjugate
sum_z = z1 + z2
conj_sum = sum_z.conjugate()

# Calculate sum of conjugates
sum_of_conj = z1.conjugate() + z2.conjugate()

print(f"z₁ = {z1}")
print(f"z₂ = {z2}")
print(f"z₁ + z₂ = {sum_z}")
print()
print(f"(z₁ + z₂)* = {conj_sum}")
print(f"z₁* + z₂* = {z1.conjugate()} + {z2.conjugate()} = {sum_of_conj}")
print(f"Equal? {conj_sum == sum_of_conj}")

print()
print("Mathematical proof:")
print("Let z₁ = a + bj and z₂ = c + dj")
print("z₁ + z₂ = (a + c) + (b + d)j")
print("(z₁ + z₂)* = (a + c) - (b + d)j")
print("           = (a - bj) + (c - dj)")
print("           = z₁* + z₂*")

# Test with multiple random complex numbers to verify
print("\n" + "="*50)
print("=== Verification with Random Complex Numbers ===")

for i in range(5):
    # Generate random complex numbers
    real1, imag1 = np.random.randn(2)
    real2, imag2 = np.random.randn(2)
    z1_test = complex(real1, imag1)
    z2_test = complex(real2, imag2)
    
    # Test property 1
    product_test = z1_test * z2_test
    prop1_check = np.isclose(abs(product_test), abs(z1_test) * abs(z2_test))
    
    # Test property 2
    sum_test = z1_test + z2_test
    prop2_check = (sum_test.conjugate() == z1_test.conjugate() + z2_test.conjugate())
    
    print(f"Test {i+1}: Property 1: {prop1_check}, Property 2: {prop2_check}")

print("\nBoth properties hold for all tested complex numbers!")
```

## Bonus: Additional Complex Number Applications

Let's explore some additional applications and properties of complex numbers that demonstrate their power in mathematics and engineering.

```python
# Bonus Applications

import cmath

print("=== Application 1: Solving Quadratic Equations ===")
print()

def solve_quadratic(a, b, c):
    """Solve ax² + bx + c = 0 using complex numbers"""
    discriminant = b**2 - 4*a*c
    sqrt_discriminant = cmath.sqrt(discriminant)
    
    x1 = (-b + sqrt_discriminant) / (2*a)
    x2 = (-b - sqrt_discriminant) / (2*a)
    
    return x1, x2

# Example: x² + x + 1 = 0 (no real solutions)
a, b, c = 1, 1, 1
x1, x2 = solve_quadratic(a, b, c)
print(f"Equation: {a}x² + {b}x + {c} = 0")
print(f"Solutions: x₁ = {x1}, x₂ = {x2}")
print(f"Verification: x₁² + x₁ + 1 = {x1**2 + x1 + 1}")

print("\n" + "="*50)
print("=== Application 2: Rotations in the Complex Plane ===")
print()

def rotate_point(z, angle_degrees):
    """Rotate a complex number by given angle"""
    angle_radians = np.radians(angle_degrees)
    rotation = cmath.exp(1j * angle_radians)
    return z * rotation

# Rotate a point through various angles
point = 1 + 0j  # Point on real axis
angles = [0, 45, 90, 135, 180, 225, 270, 315]

plt.figure(figsize=(8, 8))
for i, angle in enumerate(angles):
    rotated = rotate_point(point, angle)
    plt.plot(rotated.real, rotated.imag, 'o', markersize=8, 
             label=f'{angle}°: {rotated:.2f}')
    plt.arrow(0, 0, rotated.real, rotated.imag, 
              head_width=0.05, head_length=0.05, fc='blue', ec='blue', alpha=0.5)

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title('Rotation of Complex Number 1+0j')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

print("Complex number multiplication by e^(iθ) rotates by angle θ")

print("\n" + "="*50)
print("=== Application 3: Signal Processing (Fourier Transform) ===")
print()

# Simple demonstration of complex exponentials in signal processing
t = np.linspace(0, 2*np.pi, 1000)
frequency = 2  # Hz

# Complex exponential: e^(i*2πft)
complex_signal = np.exp(1j * 2 * np.pi * frequency * t)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(t, complex_signal.real, 'b-', label='Real part')
plt.title('Real Part of e^(i2πft)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(t, complex_signal.imag, 'r-', label='Imaginary part')
plt.title('Imaginary Part of e^(i2πft)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(complex_signal.real, complex_signal.imag, 'g-', alpha=0.7)
plt.title('Complex Plane Representation')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.axis('equal')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(t, np.abs(complex_signal), 'm-', label='Magnitude')
plt.plot(t, np.angle(complex_signal), 'c-', label='Phase')
plt.title('Magnitude and Phase')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

print("Complex exponentials are fundamental in signal processing and Fourier analysis")
print("They represent sinusoidal signals with both magnitude and phase information")
```
