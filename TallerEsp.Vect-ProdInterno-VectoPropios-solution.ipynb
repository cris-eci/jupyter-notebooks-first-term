### Exercise 2: Complex Internal Product for Square Complex Matrices
Given two matrices:
$A = \begin{bmatrix} 1+i & 2-2i \\ 3+3i & 4-i \end{bmatrix}$, $B = \begin{bmatrix} 1-2i & 3+i \\ 4-4i & 2+2i \end{bmatrix}$
Calculate $\langle A, B \rangle = \text{Tr}(\overline{A}^T B)$.

```python
# Exercise 1 Solution
a = np.array([1 + 2j, 3 - 4j])
b = np.array([2 - 1j, -1 + 3j])
internal_product = np.vdot(a, b)
print('a:', a)
print('b:', b)
print('Internal product <a, b>:', internal_product)
```

### Exercise 2: Eigenvalues and Eigenvectors of a Complex Matrix
Matrix: $B = [[1+2i, 2+3i], [4+5i, 6+7i]]$

```python
# Exercise 2 Solution
B = np.array([[1+2j, 2+3j], [4+5j, 6+7j]], dtype=complex)
eigvals_B, eigvecs_B = np.linalg.eig(B)
print('Eigenvalues:', eigvals_B)
print('Eigenvectors:', eigvecs_B)
```

### Exercise 3: Symmetric Matrix Eigendecomposition
Matrix: $C = [[5, 4], [4, 5]]$

```python
# Exercise 3 Solution
C = np.array([[5, 4], [4, 5]])
eigvals_C, eigvecs_C = np.linalg.eig(C)
print('Eigenvalues:', eigvals_C)
print('Eigenvectors:', eigvecs_C)
print('Note: For symmetric matrices, eigenvalues are real and eigenvectors are orthogonal.')
```

### Exercise 4: Plotting Matrix Transformations
Matrix: $T = [[2, 1], [1, 3]]$

```python
# Exercise 4 Solution
T = np.array([[2, 1], [1, 3]])
theta = np.linspace(0, 2*np.pi, 100)
unit_circle = np.array([np.cos(theta), np.sin(theta)])
transformed = T @ unit_circle
eigvals_T, eigvecs_T = np.linalg.eig(T)
plt.figure(figsize=(8,8))
plt.plot(unit_circle[0], unit_circle[1], 'b', label='Unit Circle')
plt.plot(transformed[0], transformed[1], 'r', label='Transformed')
# Plot eigenvectors before and after transformation
for i in range(2):
    v = eigvecs_T[:, i]
    plt.arrow(0, 0, v[0], v[1], color='g', width=0.03, label=f'Eigenvector {i+1}')
    tv = T @ v
    plt.arrow(0, 0, tv[0], tv[1], color='m', width=0.03, label=f'Transformed Eigenvector {i+1}')
plt.axis('equal')
plt.legend()
plt.title('Matrix Transformation and Eigenvectors')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
print('Eigenvalues:', eigvals_T)
print('Eigenvectors:', eigvecs_T)
```

### Exercise 5: Diagonal Matrix Eigendecomposition
Matrix: $D = [[7, 0], [0, -3]]$

```python
# Exercise 5 Solution
D = np.array([[7, 0], [0, -3]])
eigvals_D, eigvecs_D = np.linalg.eig(D)
print('Eigenvalues:', eigvals_D)
print('Eigenvectors:', eigvecs_D)
print('Note: For diagonal matrices, eigenvalues are the diagonal elements and eigenvectors are the standard basis.')
```
